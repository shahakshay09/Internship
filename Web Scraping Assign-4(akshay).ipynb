# Q.1)  Scrape the details of most viewed videos on YouTube from Wikipedia

```python
url =  'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver=webdriver.Chrome(r"C:\Users\shaha\Downloads\chromedriver_win32/chromedriver.exe")
driver.get(url)
```

```python
rank =[]
name =[]
artist = []
views = []
date = []
```

```python
for i in range(1,31):
    num_tag=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[1]")
    try:
      rank.append(num_tag.text)
    except NoSuchElementException:
      rank.append("-")
       
for i in range(1,31):
    vid_tag=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[2]")
    try:
        name.append(vid_tag.text.strip().split('"')[1]) #appending only the names of the videos
    except NoSuchElementException:
        name.append("-")
 
for i in range(1,31):
    uploadr_tag=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[3]")
    try:
        artist.append(uploadr_tag.text)
    except NoSuchElementException:
        artist.append("-")
        
for i in range(1,31):
    views_tag=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[4]")
    try:
        views.append(views_tag.text)
    except NoSuchElementException:
        views.append("-")
        
        
for i in range(1,31):
    date_tag=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[5]")
    try:
        date.append(date_tag.text)
    except NoSuchElementException:
        date.append("-")

top_youtube_vids = {"Rank":rank,"Name":name,"Artist":artist,"Upload Date":date,"Views(Billions)":views}
Most_Viewed_Videos = pd.DataFrame(top_youtube_vids)

```

```python
#scraping Last week rank
last_week=[]
try:
    L_tag=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
    for i in L_tag:
        last_week.append(i.text)
except NoSuchElementException:#handling no such element exception
    last_week.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    last_week.append('No details available')
```

```python
peak_rank=[]
#scraping peak rank
try:
    p_tag=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
    for i in p_tag:
        peak_rank.append(i.text)
except NoSuchElementException:
    peak_rank.append('No details available')
except StaleElementReferenceException:
    peak_rank.append('No details available')
```

```python
week=[]
#scraping weeks on board
try:
    w_tag=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
    for i in w_tag:
        week.append(i.text)
except NoSuchElementException:
    week.append('No details available')
except StaleElementReferenceException:
    week.append('No details available')
```

```python
df = pd.DataFrame({'song name':name,'Artist_name':artist,'Last week rank':last_week,'Peak rank':peak_rank,'Weeks on board':week})
Top_100_Songs=df[:100]
Top_100_Songs
```
